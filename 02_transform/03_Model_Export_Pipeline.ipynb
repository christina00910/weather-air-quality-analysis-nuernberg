# 📦 Finaler Modell-Fit & App-Schnittstelle (Modul 3)

## 📌 Genereller Fahrplan 
* **Finales Full-Fit:** Das Gewinner-Modell aus Modul 2 wird ein allerletztes Mal auf **100% aller verfügbaren Daten (1980–2024)** trainiert. Es gibt hier keinen Test-Split mehr.
* **Modell-Serialisierung:** Export des trainierten Modells mittels `joblib.dump()` als `wetter_schadstoff_model.pkl` in den Ordner `/models`.
* **Dummy-Schnittstellentest für Frontend:**
    * Laden der `.pkl`-Datei.
    * Erstellen eines fiktiven Arrays (z. B. Werte für Temperatur, Wind, Monat, Stunde) und Testen der `.predict()`-Funktion, um sicherzustellen, dass die App später fehlerfrei läuft.
