# 🤖 Modell-Training & Validierungs-Strategie (Modul 2)

## 📌 Genereller Fahrplan für Christina & Markus
* **Daten-Import:** Einlesen der `train_ready_data.csv`.
* **Definition von Features (X) und Target (y):** * `y` = Schadstoffkonzentration (z. B. `PM10` oder `Ozon`).
    * `X` = Wetterdaten, Zeit-Features, Lags und Wind-Features.
* **Chronologischer Train/Test-Split (⚠️ Extrem Wichtig):**
    * **KEIN** zufälliger Split (`train_test_split`), da es sich um eine Zeitreihe handelt!
    * *Vorschlag:* Trainingsdaten = 1980 bis 2020. Testdaten = 2021 bis 2024. Wir evaluieren, wie gut das Modell die "Zukunft" vorhersagt.

## 📊 ML-Modelle im Vergleich
1.  **Baseline-Modell:** Eine einfache `LinearRegression`, um eine Untergrenze für die Metriken zu haben.
2.  **Komplexes Baum-Modell:** `RandomForestRegressor` oder `XGBoost`. Diese können komplexe, nicht-lineare Zusammenhänge (wie Temperatur vs. Ozon) perfekt abbilden.
3.  **Evaluation:** Berechnung von **RMSE** (Abweichung in µg/m³), **MAE** und **R²** auf dem zeitlich getrennten Test-Set zur Ermittlung des "Gewinner-Modells".
