# 🗺️ Projekt-Fahrplan & Daten-Transformation (Modul 1)

## 📌 Genereller Fahrplan 
* **Rohdaten-Import:** Einlesen der `Schadstoff_Wetter.csv`.
* **Umgang mit fehlenden Werten (NaNs):** Wetterdaten lückenlos machen. **Wichtig:** Keine Zeilen löschen (`dropna()`), da sonst die Zeitreihe bricht! Stattdessen Vorwärts-Auffüllen (`ffill()`) oder lineare Interpolation nutzen.
* **Zeitstempel-Harmonisierung:** Zusammenführen von `datum` und `stunde` in ein echtes Pandas-Datetime-Objekt.
* **Extraktion zyklischer Zeit-Features:** Erstellen der Spalten `Monat`, `Wochentag` und `Stunde`.
* **Lag-Features (Verzögerungen):** Erstellen von Prädiktoren wie `Ozon_vor_24h` oder `PM10_vor_1h`.

## 🌬️ Integration der Wind-Dispersion
* **Meteorologischer Kontext:** Starker Wind verdünnt Schadstoffe (Dispersion). Flaute sorgt für Schadstoffstau.
* **Feature-Engineering:** * `windgeschwindigkeit` unverändert als Feature behalten.
    * Erstellung eines gleitenden Durchschnitts (Rolling Window), z. B. `Wind_Schnitt_letzte_6h`, um langanhaltende Wetterlagen zu erfassen.
* **Daten-Export:** Speichern des bereinigten, erweiterten Datensatzes als `train_ready_data.csv` im Ordner `/data`.
